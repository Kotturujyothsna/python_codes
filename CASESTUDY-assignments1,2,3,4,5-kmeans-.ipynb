```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
df = pd.read_csv('customers.csv')
df
```

```python
print("\nNull Values in Each Column:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Spending Score (1-100)'], c='blue', label='Data Points')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title('Scatter Plot of Age vs. Spending Score')
plt.legend()
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, n_init=25, random_state=42)
kmeans
```

```python
X = df[['Age', 'Spending Score (1-100)']]
df['Cluster'] = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Spending Score (1-100)', hue='Cluster', palette='viridis', s=100, alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title('Scatter Plot of Age vs. Spending Score with Clusters')
plt.legend(title='Cluster')
plt.show()
```

```python
df = pd.read_csv('customers.csv')
```

```python
print("\nNull Values in Each Column:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Spending Score (1-100)'], c='blue', label='Data Points')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title('Scatter Plot of Age vs. Spending Score')
plt.legend()
plt.show()
```

```python
X = df[['Age', 'Spending Score (1-100)']]
```

```python
X
```

```python
sse = []
```

```python
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, n_init=25, random_state=42)
    kmeans.fit(X)
    sse.append(kmeans.inertia_) 
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Distances (SSE)')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()
```

```python
optimal_clusters = 5
```

```python
kmeans = KMeans(n_clusters=optimal_clusters, n_init=25, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
print("\nCluster Centers:")
print(kmeans.cluster_centers_)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Spending Score (1-100)', hue='Cluster', palette='viridis', s=100, alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title(f'Scatter Plot of Age vs. Spending Score with {optimal_clusters} Clusters')
plt.legend(title='Cluster')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

data = np.array([
    [5, 3],
    [10, 15],
    [15, 12],
    [24, 10],
    [30, 30],
    [85, 70],
    [71, 80],
    [60, 78],
    [70, 55],
    [80, 91]
])
data
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c='blue', marker='o', s=100, alpha=0.7)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of the Data')
plt.grid(True)
plt.show()
```

```python
linked = linkage(data, method='ward')
linked
```

```python
plt.figure(figsize=(10, 7))
dendrogram(linked, 
           orientation='top', 
           distance_sort='descending',
           show_leaf_counts=True)
plt.title('Dendrogram for Agglomerative Clustering')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```
