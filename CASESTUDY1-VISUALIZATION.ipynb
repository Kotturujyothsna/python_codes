```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('customer_churn-1.csv')


plt.figure(figsize=(10, 6))
df['Contract'].value_counts().plot(kind='bar', color='orange')


plt.xlabel('Contract Type of customer')
plt.ylabel('Count')
plt.title('Distribution of Contract')

plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.hist(df['MonthlyCharges'], bins=30, color='forestgreen', edgecolor='black')

plt.xlabel('Monthly Charges Incurred')
plt.ylabel('Count')
plt.title('Distribution of Monthly Charges')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('customer_churn-1.csv')

plt.figure(figsize=(10, 6))
plt.scatter(df['tenure'], df['TotalCharges'], color='indigo')

plt.xlabel('Tenure of the customer')
plt.ylabel('Total charges Incurred')
plt.title('Total Charges vs Tenure')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('customer_churn-1.csv')

plt.figure(figsize=(12, 6))
df.boxplot(column='MonthlyCharges', by='PaymentMethod', patch_artist=True, 
           boxprops=dict(facecolor='olive'))


plt.xlabel('Payment Method of customer')
plt.ylabel('Monthly Charges Incurred')
plt.title('Monthly Charges vs. Payment Method')
plt.suptitle('') 
plt.show()
```
