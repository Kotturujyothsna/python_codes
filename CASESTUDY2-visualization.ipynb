```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('cars-1.csv')

plt.figure(figsize=(10, 6))
df['cyl'].value_counts().sort_index().plot(kind='bar', color='skyblue', edgecolor='black')


plt.xlabel('Number of cylinders')
plt.ylabel('Frequency of cars')
plt.title('Number of Cylinders vs Frequency of Cars')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('cars.csv')

if df.empty:
    print("The DataFrame is empty. Please check the data source.")
else:
   
    null_counts = df.isnull().sum()

    null_counts = null_counts[null_counts > 0]

    if null_counts.empty:
        print("No columns with null values found.")
    else:
       
        print("Columns with null values and their counts:")
        print(null_counts)

        plt.figure(figsize=(10, 6))
        null_counts.plot(kind='bar', color='coral', edgecolor='black')

        plt.xlabel('Columns')
        plt.ylabel('Number of Null Values')
        plt.title('Count of Null Values in Each Column')

        plt.show()

```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['mpg'], bins=30, color='lightgreen', edgecolor='black')
plt.xlabel('Miles per gallon')
plt.ylabel('Frequency')
plt.title('Miles Per Gallon Histogram')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df['hp'], patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.xlabel('Car Horsepower')
plt.title('Boxplot for Car Horsepower')
plt.show()
```
