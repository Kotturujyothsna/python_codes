```python
import pandas as pd

df = pd.read_csv('cars-1.csv')
print(df.columns)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python

plt.figure(figsize=(10, 6))
plt.plot(df['model'], df['hp'], marker='o', linestyle='-', color='b')


plt.xlabel('Models of the cars')
plt.ylabel('Horse-Power of Cars')
plt.title('Model Names vs HorsePower')


plt.xticks(rotation=90)


plt.show()

```

```python

plt.figure(figsize=(10, 6))

plt.bar(df['carb'], df['gear'], color='skyblue')

plt.xlabel('Number of carburetors')
plt.ylabel('Number of forward gears')
plt.title('Carbs vs Gear')


plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.hist(df['wt'], bins=30, color='skyblue', edgecolor='black')

plt.xlabel('Weight of the cars')
plt.ylabel('Count')
plt.title('Histogram for the Weight Values')

plt.show()
```

```python
df.head()
```

```python

cyl_counts = df['cyl'].value_counts()
```

```python

plt.figure(figsize=(8, 8))
plt.pie(cyl_counts, labels=cyl_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.get_cmap('Pastel1').colors)

plt.title('Distribution of Car Models by Number of Cylinders')

plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.fill_between(df['am'], df['carb'], color='skyblue', alpha=0.4)

plt.xlabel('Transmission')
plt.ylabel('Number of carburetors')
plt.title('Transmission vs Number of Carburetors')


plt.show()
```
