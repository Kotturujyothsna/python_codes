```python
import pandas as pd
import numpy as np

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data'
column_names = [
    'age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num'
]
df = pd.read_csv(url, header=None, names=column_names, na_values='?', delimiter=',')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# Drop rows with missing target values
df.dropna(subset=['num'], inplace=True)
```

```python
# Fill missing values in other columns
imputer = SimpleImputer(strategy='mean')
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
```

```python
df_imputed
```

```python
# Separate features and target variable
X = df_imputed.drop(columns=['num', 'has_disease'])
y = df_imputed['has_disease']

```

```python
X
```

```python
y
```

```python
# Split the data into training and test sets (70:30 ratio)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python

# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
y_pred
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)

# Calculate the accuracy score
accuracy = accuracy_score(y_test, y_pred)
```

```python
conf_matrix
```

```python
accuracy
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn import tree

# Initialize and train the Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train_scaled, y_train)
```

```python
# Predict on the test set
dt_y_pred = dt_model.predict(X_test_scaled)
dt_y_pred
```

```python
# Calculate the confusion matrix
dt_conf_matrix = confusion_matrix(y_test, dt_y_pred)

# Calculate the accuracy score
dt_accuracy = accuracy_score(y_test, dt_y_pred)
```

```python
dt_conf_matrix
```

```python
dt_accuracy
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize and train the Random Forest model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_scaled, y_train)
```

```python

# Predict on the test set
rf_y_pred = rf_model.predict(X_test_scaled)
```

```python
rf_y_pred
```

```python
# Calculate the confusion matrix
rf_conf_matrix = confusion_matrix(y_test, rf_y_pred)

# Calculate the accuracy score
rf_accuracy = accuracy_score(y_test, rf_y_pred)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Logistic Regression
lr_precision = precision_score(y_test, y_pred)
lr_recall = recall_score(y_test, y_pred)
lr_f1 = f1_score(y_test, y_pred)

# Decision Tree
dt_precision = precision_score(y_test, dt_y_pred)
dt_recall = recall_score(y_test, dt_y_pred)
dt_f1 = f1_score(y_test, dt_y_pred)

# Random Forest
rf_precision = precision_score(y_test, rf_y_pred)
rf_recall = recall_score(y_test, rf_y_pred)
rf_f1 = f1_score(y_test, rf_y_pred)

print(f"\nLogistic Regression - Precision: {lr_precision}, Recall: {lr_recall}, F1 Score: {lr_f1}")
print(f"Decision Tree - Precision: {dt_precision}, Recall: {dt_recall}, F1 Score: {dt_f1}")
print(f"Random Forest - Precision: {rf_precision}, Recall: {rf_recall}, F1 Score: {rf_f1}")
```

```python
def plot_confusion_matrix(cm, title='Confusion Matrix'):
    plt.figure(figsize=(10, 7))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False, 
                xticklabels=['No Disease', 'Disease'], yticklabels=['No Disease', 'Disease'])
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()
```

```python
plot_confusion_matrix(conf_matrix, title='Logistic Regression Confusion Matrix')
plot_confusion_matrix(dt_conf_matrix, title='Decision Tree Confusion Matrix')
plot_confusion_matrix(rf_conf_matrix, title='Random Forest Confusion Matrix')
```
